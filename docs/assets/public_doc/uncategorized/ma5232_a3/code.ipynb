```python
import math # import a package
import numpy as np # import a package under an alias
from sklearn import linear_model # import a submodule
from os import mkdir # import a specific function
```

```python
from sklearn.datasets import make_blobs
n_samples = 50
n_features = 2
offset = 0.5
seed = 42
centers = np.array ([
    [0 , 0],
    [offset, 0],
])
Z, c = make_blobs(n_samples=n_samples, centers=centers, n_features=n_features, random_state=seed, cluster_std=0.05, shuffle=False)
X = Z[c == 0] # source distribution
Y = Z[c == 1] # target distribution
```

```python
def plot2D_samples_mat(xs, xt, G, ax, thr=1e-8, **kwargs):
    if ("color" not in kwargs) and ("c" not in kwargs):
        kwargs["color"] = "k"
    mx = G.max()
    if "alpha" in kwargs:
        scale = kwargs["alpha"]
        del kwargs["alpha"]
    else:
        scale = 1

    for i in range(xs.shape[0]):
        for j in range(xt.shape[0]):
            if G[i, j] / mx > thr:
                ax.plot([xs[i, 0], xt[j, 0]],
                        [xs[i, 1], xt[j, 1]],
                        alpha=G[i, j] / mx*scale, **kwargs,
                )
```

```python
# import POT, plt
import ot
from matplotlib import pyplot as plt
# eps
eps = 1e-6
```

```python
# get data shape - move n points into m points
n, d = X.shape
m, d = Y.shape
m, n, d
```

## (a) Compute the two cost matrices between samples:
- $d_1 = (||x_i - y_i||_2)_{ij}$
- $d_2 = (||x_i - y_i||_2^2)_{ij}$

```python
d1 = ot.dist(X, Y, metric="euclidean")
d2 = ot.dist(X, Y, metric="sqeuclidean")

assert tuple(d1.shape) == tuple(d1.shape) == (n, m)
assert np.max(np.abs(d1**2 - d2)) < eps
```

## (b) Computer the weights of each point in the distribution (use uniform weights)

$$
    w_x, w_y
$$

```python
wx = np.ones(shape=(n,)) / n
wy = np.ones(shape=(m,)) / m

assert np.sum(wx) - 1 < eps
assert np.sum(wy) - 1 < eps
```

## (c) For each cost matrix, compute the optimal transport plan between the source and target distributions

```python
T1 = ot.emd(wx, wy, d1)
T2 = ot.emd(wx, wy, d2)
```

## (d) Plot the two optimal transport plan along with the source and target samples

```python
fig, axes = plt.subplots(ncols=2)
ax1, ax2 = axes
ax1.set_title("l2 norm")
ax2.set_title("l2 norm squared")
plot2D_samples_mat(X, Y, T1, ax=ax1)
plot2D_samples_mat(X, Y, T2, ax=ax2)
plt.show()
```

## (e) What is the major difference between the two optimal transport plans?

There are many _crossings_ in $l_2$-norm squared distance.

Informally speaking, transporting between two uniform distributions supported on finite sets $X, Y$ of the same size $|X| = n = m = |Y|$ consists of _basic transport_, that is, how much Earth to move from $x \in X$ into $y \in Y$. When plotting, each _basic transport_ corresponds to a line segment connecting $X$ to $Y$.

Since the maximum distance between any two points $x \in X$ and $y \in Y$ is less than $1$, informally speaking, $l_2$-norm squared distance put less penalty, so it enables the transport to travel further with respect to Euclidean distance to achieve optimal transport. Hence, there are more _crossings_

## (f) What are the Wasserstein distances in these two cases?

```python
W1 = ot.emd2(wx, wy, d1)
W2 = ot.emd2(wx, wy, d2)

assert W1 - np.sum(T1 * d1) < eps
assert W2 - np.sum(T2 * d2) < eps

W1, W2
```
